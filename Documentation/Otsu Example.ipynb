# `Otsu` Example 1

Importing modules.

```python
from cv import Otsu
```

Initializing `Otsu` object.

```python
otsu = Otsu()
```

Using `load_image()` function to load the image that we are going to process.

```python
otsu.load_image('dataset_test/map.jpg')
```

How the image looks like.

```python
otsu.show_original_image()
```

We need to call the `create_histogram()` method prior to calculating the best threshold value. This might take a while as it iterates through all pixels in the image.

```python
otsu.create_histogram()
```

How the histogram of the image above looks like.

```python
otsu.show_histogram()
```

We will use `find_optimum_threshold()` to find out the best threshold to perform binarization. The function is going to check every single intensity level (0 to 255) and choose the one with the least Within Class Variance (WCV) value. In this case the most optimum threshold is 157 with the WCV of 331.455.

```python
otsu.find_optimum_threshold()
```

As the optimum threshold has been found, now we can call the `show_thresholded_image()` function to display the binarization result. By running the code below, all dark pixels (in this case <157) are going to be rounded down to 0, while the rest (>=157) are going to be set to 1.

```python
otsu.show_thresholded_image()
```

# `Otsu` Example 2

Another example: using histogram only (no image input).
Assume that we have a histogram that looks like the following.

Note: intensity level 0: 8 pixels, intensity level 1: 7 pixels, intensity level 2: 2 pixels, and so forth.

```python
histogram = [8,7,2,6,9,4]
```

Initialize another `Otsu` object.

```python
otsu2 = Otsu()
```

Load the histogram that I created manually above using `load_histogram()` function.

```python
otsu2.load_histogram(histogram)
```

Now we can see that the optimum threshold for the histogram is 3, with the WCV of 0.491.

```python
otsu2.find_optimum_threshold()
```

Keep in mind that since we don't work with any image on `otsu2`, hence both `show_thresholded_image()` and `show_original_image()` will not work.

```python
otsu2.show_original_image()
```

```python
otsu2.show_thresholded_image()
```
