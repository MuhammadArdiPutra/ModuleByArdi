# `Convolution` Example

Import the `Convolution` module.

```python
from Ardi.cv import Convolution
```

Initialize a `Convolution` object with random kernel.

```python
convolution = Convolution(kernel='random', random_state=77)
```

Below is how the kernel that we are going to apply to an image looks like.

```python
convolution.kernel
```

Load the image using `load_image()` function. Here I also reduce the image resolution by 90% (by setting the `resolution` parameter to 0.1).

```python
convolution.load_image('dataset_test/flower.jpg', resolution=0.1)
```

Displaying the original image in grayscale using `show_original()` function.

```python
convolution.show_original()
```

Displaying the resized image using `show_resized()` function. Notice that even though the image look almost exactly the same, yet the xticks and yticks values emphasizes that the resized image is way smaller than the original image. In fact, such a reduction in image dimension makes the convolution process more effective (since our kernel is relatively small as well).

```python
convolution.show_resized()
```

Displaying the zero-padded image using `show_zero_padded()` function. The resulting image might look very similar to the one obtained using `show_resized()` function because the zero-padding only gives an additional dimension of 1 on each side. This basically means that if you take a very close look to the image, then there will be an extremely thin black frame around the image.

```python
convolution.show_zero_padded()
```

Start the convolution process by calling `convolve()` method. Note that the value displayed in the progress bar below is the image height.

```python
convolution.convolve()
```

Below is how the convolved image looks like.

```python
convolution.show_convolved()
```

Below is another example. The steps are exactly the same except the kernel type.

```python
convolution2 = Convolution(kernel='vertical_edge')
convolution2.load_image('dataset_test/flower.jpg', resolution=0.1)
convolution2.convolve()
convolution2.show_convolved()
```
