# ```Perceptron_2D``` Example

Importing modules.

````python
from Ardi.ml import Perceptron_2D
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
````

Defining datapoints using Sklearn. You may also create the samples manually like below.

````python
X, y = make_blobs(n_samples=1000, centers=2, n_features=2, cluster_std=0.5, random_state=0)
plt.scatter(X[:,0], X[:,1], c=y, cmap='winter')
````

```Perceptron_2D``` object initialization with 0.2 learning rate and 100 iterations.

````python
perceptron = Perceptron_2D(learning_rate=0.2, iterations=100, random_state=6)
````

We use ```take_data_raw()``` instead of ```take_data_csv()``` since the features and samples are already stored in different arrays (```X``` and ```y```).

````python
perceptron.take_data_raw(X, y)
````

Finding out the number of samples by taking ```n_samples``` attribute.

````python
perceptron.n_samples
````

Finding out the number of features by taking the ```n_features``` attribute.

````python
perceptron.n_features
````

Start the training process.

````python
perceptron.train()
````

````python
print('Old weight_0\t:', perceptron.weight_0)
print('Old weight_1\t:', perceptron.weight_1)
print('Old bias\t:', perceptron.bias)
````

````python
print('Updated weight_0\t:', perceptron.updated_weight_0)
print('Updated weight_1\t:', perceptron.updated_weight_1)
print('Updated bias\t\t:', perceptron.updated_bias)
````

Predicting multiple samples.

````python
perceptron.predict_multiple_samples([[2,20], [2,2], [3,3]])
````

Display training progress.

````python
perceptron.plot_errors(print_details=True)
````

Displaying the decision boundary before training.

````python
perceptron.plot_before(xlabel='Feature 1', ylabel='Feature 2')
````

Displaying the decision boundary after training.

````python
perceptron.plot_after(xlabel='Feature 1', ylabel='Feature 2')
````
